```python
dimension = 20
with open('_evals-%dD.py' % dimension, 'rt') as file:
    evals = ast.literal_eval(file.read())
with open('_stops-%dD.py' % dimension, 'rt') as file:
    stops = ast.literal_eval(file.read())
```

```python
factors = sorted(evals)
# gcf().clear()
medians = []
for factor in factors:
    data = evals[factor]  # an alias
    print('%.2f:%d' % (factor, len(data)), end=' ')
    style = 'k*'
    if not all([s == {'ftarget': 1e-8} for s in stops[factor]]):
        style = 'r+'
    loglog(len(data) * [factor], data, style)
    medians += [median(data)]
loglog(factors, medians)
text(factors[0], medians[0], '%d-D' % dimension)
xlabel('factor'); ylabel('f-evaluations'); grid(True, 'minor');
```
